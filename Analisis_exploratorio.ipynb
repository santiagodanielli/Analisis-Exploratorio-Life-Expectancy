```python
# Importar las librerías para creación de dataframe, análisis y visualización
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px

# Crear variable-objeto dataframe con csv y ver su estructura

df = pd.read_csv("LifeExpectancy.csv")

df.shape
# Para conocer la estructura tambien se pueden utilizar head, info, describe
```

```python
# Conocer las columnas, la opción tolist() muestra el listado completo, no importa la cantidad de columnas

df.columns.tolist()
```

```python
# Conocer el tipo de dato que tiene una columna

df.dtypes
```

```python
# Renombrar una columna

df = df.rename(columns={"Country": "Pais"})

df
```

```python
# Renombrar más de una columna

df = df.rename(columns={"Year": "Año",
                        "Status": "Estado pais",
                        "Life expectancy": "Esperanza de vida"})

df
```

```python
# Comenzar la exploración de los datos. 
# ¿Qué cantidad de datos perdidos hay en cada variable?
df.isnull().sum()
```

```python
# ¿Que cantidad de categorías tiene la variable "estado país"?
# ¿Como se distribuyen las frecuencias?

df["Estado pais"].value_counts()
```

```python
# Todas las medidas de tendencia central de la variable "esperanza de vida"

df["Esperanza de vida"].describe()
```

```python
# ¿Cual es promedio de esperanza de vida en países en vías de desarrollo y desarrollados?

df["Esperanza de vida"].groupby([df["Estado pais"]]).mean()
```

# 3. Análisis univariado variables cualitativas. Gráficos.

```python
# Visualizar los datos de la celda previa con seaborn
# ¿Como se distribuyen las frecuencias de la variable "estado país"?
sns.countplot(x = "Estado pais",
              data = df)
```

```python
# Visualizar sobre el eje y
# Agregar título y tamaño de figura

plt.figure(figsize=(10,6))
plt.title("Cantidad de paises segun su estado de desarrollo")
sns.countplot(y = "Estado pais",
              data = df)
```

# 4. Análisis univariado variables cuantitativas. Gráficos.

```python
# Histograma opcion integrada de pandas.

df["Esperanza de vida"].hist(bins=10, 
                             grid=False, 
                             edgecolor="black")                             
```

```python
# Histograma con seaborn. Con el menor código posible nos muestra una visualización mucho más agradable

sns.histplot(x=df["Esperanza de vida"])
```

```python
# Agregando o modificando características como tamaño, título y color.

plt.figure(figsize=(10,6))
plt.title("Distribución de frecuencias esperanza de vida")
sns.histplot(x=df["Esperanza de vida"],
             color="pink")
```

```python
# Boxplot sobre la variable estatura con seaborn

plt.figure(figsize=(10,6))
plt.title("Medidas tendencia central Esperanza de vida")
sns.boxplot(y=df["Esperanza de vida"])
```

```python
# Explorando plotly. A diferencia de seaborn, plotly es interactivo

px.box(y=df["Esperanza de vida"])
```

# 5. Análisis bivariado variables cualitativas. Gráficos.

```python
# Esperanza de vida y estado de los países (en este caso es una combinación entre una cualitativa y una cuantitativa)

plt.figure(figsize=(10,6))
plt.title("Distribución de frecuencias esperanza de vida")
sns.histplot(x=df["Esperanza de vida"],
             color="pink",
             hue=df["Estado pais"])
```

# 6. Análisis bivariado variables cuantitativas. Gráficos.



```python
# Gráfico de dispersión tasa de homicidios y porcentaje de pobreza

sns.scatterplot(x=df["Schooling"],
                y=df["Esperanza de vida"])
```

```python
# Correlación entre variables y heat map
# Creación de la variable matriz de correlación para seleccionar que variables se quieren correlacionar

matriz_corr = df[["Esperanza de vida",
                  "Schooling",
                  "GDP",
                  "Total expenditure"]].corr()

# Heatmap
plt.figure(figsize=(10,6))
plt.title("Matriz de correlación")
sns.heatmap(matriz_corr,
            annot=True,
            cmap="coolwarm",
            center=0)
```
